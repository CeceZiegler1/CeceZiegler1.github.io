```python
import numpy as np

def convolve(X, K):
    m, n = X.shape
    k = K.shape[0]
    
    pad = (k - 1) // 2
    
    # Create a zero-padded version of the input image
    X_pad = np.pad(X, pad, mode='constant', constant_values=0)
    
    # Initialize the output image
    Y = np.zeros((m, n))
    
    for i in range(m):
        for j in range(n):
            # Extract the local neighborhood of the current pixel from the padded input image
            neighborhood = X_pad[i:i+k, j:j+k]
            
            # Compute the dot product between the neighborhood and the kernel
            Y[i,j] = np.sum(neighborhood * K)
    
    return Y
```

```python
from PIL import Image
import urllib
import numpy as np
from matplotlib import pyplot as plt

def read_image(url):
    return np.array(Image.open(urllib.request.urlopen(url)))

url = "https://i.pinimg.com/originals/0e/d0/23/0ed023847cad0d652d6371c3e53d1482.png"

img = read_image(url)

def to_greyscale(im):
    return 1 - np.dot(im[...,:3], [0.2989, 0.5870, 0.1140])

img = to_greyscale(img)

plt.imshow(img, cmap = "Greys")
plt.gca().axis("off")
```

```python
kernel = np.array([[-1, -1, -1], 
                   [-1,  8, -1], 
                   [-1, -1, -1]])

convd = convolve(img, kernel)

plt.imshow(convd, cmap = "Greys", vmin = 0, vmax = 8)
plt.gca().axis("off")
```
